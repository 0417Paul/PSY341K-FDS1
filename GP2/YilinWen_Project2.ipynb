```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
myData = pd.read_csv("wageGenderEduAge.csv")
```

```python
display(myData)
```

```python
sns.displot(myData, x="wage", hue="gender", kind="hist", alpha=0.3)
```

```python
#  This histogram shows that there are more females gets paid less than 10 dollars for their hourly wage than male.
```

```python
sns.scatterplot(data=myData, x='age', y='wage').set(title = 'Wage and Education')
#  the scatterplot shows that the wage is less when the age is younger. 
#  There are more median and high wages when the age increase.
```

```python
sns.jointplot(x="education", y="wage", data=myData, kind="reg")
#  This plot showed you will get more money if your have more education background than others. 
```

```python
g = sns.pairplot(myData, hue = "gender")
```
