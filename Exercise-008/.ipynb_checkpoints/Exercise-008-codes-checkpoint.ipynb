```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data_1 = pd.read_csv('008ExerciseFile.csv') # you can name your array whatever you want!
```

```python
display(data_1)
```

```python
sns.lmplot(data=data_1, x='rel_time', y='m_level')
#  The straight line does not fit the data too well because the residual for this plot is very huge.
```

```python
sns.residplot(data=data_1, x='rel_time', y='m_level',)
```

```python
sns.lmplot(data=data_1, x='rel_time', y='m_level', order=3)
#  This line would fit the data better due to its matching shape and reasonable residual values.
```

```python
sns.residplot(data=data_1, x='rel_time', y='m_level',order = 3)
```
